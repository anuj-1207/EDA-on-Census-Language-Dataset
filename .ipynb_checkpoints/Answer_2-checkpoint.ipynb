##### Reading language data from 'DDW-C18-0000.xlsx' file which is mentioned in Assignment 2


```python
Language_data= pd.read_excel('DDW-C18-0000.xlsx',usecols=['Unnamed: 0','Unnamed: 2','Unnamed: 3','Unnamed: 4','Unnamed: 6','Unnamed: 7','Unnamed: 9','Unnamed: 10'])
Language_data.rename(columns = {'Unnamed: 0':'Code','Unnamed: 2':'Name','Unnamed: 3':'TRU','Unnamed: 4':'Age_group',
                                'Unnamed: 6':'Males speaking 2 or more languaguages',
                                'Unnamed: 7':'Females speaking 2 or more languaguages',
                                'Unnamed: 9':'Males speaking 3 or more languaguages',
                                'Unnamed: 10':'Females speaking 3 or more languaguages',}, inplace = True)

Language_data = Language_data.drop(Language_data.index[[0,1,2,3,4]])
Language_data = Language_data.loc[(Language_data['TRU'] == 'Total') & (Language_data['Age_group'] == 'Total')]
Language_data = Language_data.reset_index()
Language_data = Language_data.drop(columns=["index","TRU","Age_group"])
Language_data = Language_data.sort_values('Code',ignore_index=True)
Language_data["Code"] = Language_data["Code"].astype(int)
Language_data.head(5)
```

```python
Processed_data = pd.merge(population_data_Total,Language_data,left_on = 'Name', right_on = 'Name')

# Processed_data['TOT_A'] = Processed_data['TOT_P'] - Processed_data['People_speaking_2_languaguages']
# Processed_data['People_speaking_2_languaguages'] -= Processed_data['People_speaking_3_languaguages']
# Processed_data.rename(columns = {'TOT_A':'No_PSE_1_language',
#                                  'People_speaking_2_languaguages':'No_PSE_2_language',
#                                 'People_speaking_3_languaguages':'No_PSE_3_language'}, inplace = True)
''' 
    No.PSE means Number of People Speaking Exactly
'''
# Processed_data.iat[0,0] = 'INDIA'
# Processed_data = Processed_data[['Name','TOT_P','No_PSE_1_language','No_PSE_2_language','No_PSE_3_language']]
Processed_data = Processed_data.sort_values('Code',ignore_index=True)
Processed_data.head(5)
```

```python
Processed_Data = Processed_data.copy()

Processed_Data['Percentage male speaking only 1 language'] = ((Processed_Data['TOT_M'] - Processed_Data['Males speaking 2 or more languaguages']) / Processed_Data['TOT_M']) * 100
Processed_Data['Percentage male speaking exactly 2 language'] = ((Processed_Data['Males speaking 2 or more languaguages'] - Processed_Data['Males speaking 3 or more languaguages']) / Processed_Data['TOT_M']) * 100
Processed_Data['Percentage male speaking 3 or more languaguages'] = ((Processed_Data['Males speaking 3 or more languaguages']) / Processed_Data['TOT_M']) * 100

Processed_Data['Percentage female speaking only 1 language'] = ((Processed_Data['TOT_F'] - Processed_Data['Females speaking 2 or more languaguages']) / Processed_Data['TOT_F']) * 100
Processed_Data['Percentage female speaking exactly 2 language'] = ((Processed_Data['Females speaking 2 or more languaguages'] - Processed_Data['Females speaking 3 or more languaguages']) / Processed_Data['TOT_F']) * 100
Processed_Data['Percentage female speaking 3 or more languaguages'] = (Processed_Data['Females speaking 3 or more languaguages'] / Processed_Data['TOT_F']) * 100

Processed_Data['ratio_1'] = (Processed_Data['TOT_M']-Processed_Data['Males speaking 2 or more languaguages'])/(Processed_Data['TOT_F']-Processed_Data['Females speaking 2 or more languaguages'])
Processed_Data['ratio_2'] = (Processed_Data['Males speaking 2 or more languaguages'] - Processed_Data['Males speaking 3 or more languaguages'])/(Processed_Data['Females speaking 2 or more languaguages']-Processed_Data['Females speaking 3 or more languaguages'])
Processed_Data['ratio_3'] = Processed_Data['Males speaking 3 or more languaguages']/Processed_Data['Females speaking 3 or more languaguages']
Processed_Data['ratio_pop'] = Processed_Data['TOT_M'] / Processed_Data['TOT_F']

Processed_Data['p-value'] = Processed_Data.apply(lambda row: scipy.stats.ttest_1samp([row.ratio_1, row.ratio_2, row.ratio_3], popmean=row.ratio_pop)[1], axis=1)

Processed_Data = Processed_Data[['Code','Name','Percentage male speaking only 1 language','Percentage male speaking exactly 2 language', 
                                 'Percentage male speaking 3 or more languaguages','Percentage female speaking only 1 language', 
                                 'Percentage female speaking exactly 2 language','Percentage female speaking 3 or more languaguages','p-value']]
                                                                       
Processed_Data.head()
```

```python
pre_final_1 = Processed_Data[['Code','Name','Percentage male speaking only 1 language','Percentage female speaking only 1 language','p-value']].copy()
pre_final_2 = Processed_Data[['Code','Name','Percentage male speaking exactly 2 language','Percentage female speaking exactly 2 language','p-value']].copy()
pre_final_3 = Processed_Data[['Code','Name','Percentage male speaking 3 or more languaguages','Percentage female speaking 3 or more languaguages','p-value']].copy()

pre_final_1.columns=['State_code','Name','male-percentage','female-percentage','p-value']
pre_final_2.columns=['State_code','Name','male-percentage','female-percentage','p-value']
pre_final_3.columns=['State_code','Name','male-percentage','female-percentage','p-value']

pre_final_1.to_csv('gender-india-a.csv')
pre_final_2.to_csv('gender-india-b.csv')
pre_final_3.to_csv('gender-india-c.csv')
```
